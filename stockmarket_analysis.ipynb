```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt


ticker = input("Enter Stock Symbol ").strip().upper()          
start_date = input("Enter Start Date (YYYY-MM-DD) ").strip()
end_date = input("Enter End Date (YYYY-MM-DD) ").strip()


print(f"\nFetching data for {ticker} from {start_date} to {end_date}\n")
df = yf.download(ticker, start=start_date, end=end_date)

if isinstance(df.columns, pd.MultiIndex):
    df.columns = [col[0] for col in df.columns]

df["MA5"] = df["Close"].rolling(5).mean()
df["MA10"] = df["Close"].rolling(10).mean()
df["Daily Return %"] = df["Close"].pct_change() * 100
df.dropna(inplace=True)

print("link for all stocks symbols available in library yfinance->https: //finance.yahoo.com/lookup/?guccounter=1&guce_referrer=aHR0cHM6Ly93d3cuZ29vZ2xlLmNvbS8&guce_referrer_sig=AQAAAB1HApeuFpxWFmRowKv0rlaWvhDr52QHStyOLDJmR3kyo8mvmwPuSwGHwNDpXQau79NVdq4GScZ7SXTumGYjMt8nRwPHqKzJBMLpYkqLeuJtpe16TSl3SooYA1_5TjrhfBv7dY4_RmleHIZ14A6qx4NKCbsfdeoxYjdWDbSsJWzehttps://finance.yahoo.com/lookup/?guccounter=1&guce_referrer=aHR0cHM6Ly93d3cuZ29vZ2xlLmNvbS8&guce_referrer_sig=AQAAAB1HApeuFpxWFmRowKv0rlaWvhDr52QHStyOLDJmR3kyo8mvmwPuSwGHwNDpXQau79NVdq4GScZ7SXTumGYjMt8nRwPHqKzJBMLpYkqLeuJtpe16TSl3SooYA1_5TjrhfBv7dY4_RmleHIZ14A6qx4NKCbsfdeoxYjdWDbSsJWze")
```

```python
plt.figure(figsize=(8,4))
plt.scatter(df["Volume"], df["Close"], alpha=0.6)
plt.title(f"{ticker} Price vs Volume")
plt.xlabel("Volume")
plt.ylabel("Close Price")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.scatter(df["Volume"], df["Daily Return %"], alpha=0.5)
plt.title(f"{ticker} Volume vs Daily Return")
plt.xlabel("Volume")
plt.ylabel("Daily Return (%)")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(df.index, df["Close"], label="Close Price", color="blue")
plt.plot(df.index, df["MA5"], label="MA 5", linestyle="--", color="orange")
plt.plot(df.index, df["MA10"], label="MA 10", linestyle="--", color="green")
plt.title(f"{ticker} Closing Price with 5 & 10 Day Moving Averages")
plt.xlabel("Date")
plt.ylabel("Price (in USD)")
plt.legend()
plt.grid(True)
```

```python
plt.figure(figsize=(10,4))
plt.plot(df.index, df["Daily Return %"], color="purple")
plt.title(f"{ticker} Daily Returns (%)")
plt.xlabel("Date")
plt.ylabel("Daily Change (%)")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.bar(df.index, df["Volume"], color="skyblue")
plt.title(f"{ticker} Trading Volume")
plt.xlabel("Date")
plt.ylabel("Volume")
plt.grid(True, axis="y", linestyle="--", alpha=0.7)
plt.show()
```
